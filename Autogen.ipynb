```python
from transformers import AutoModelForCausalLM, AutoTokenizer

# Load model and tokenizer
model_name = "gpt2"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)

def get_response(prompt):
    # Tokenize the input
    inputs = tokenizer(prompt, return_tensors="pt", truncation=True, max_length=200)
    outputs = model.generate(
        **inputs,
        max_new_tokens=50,
        pad_token_id=tokenizer.eos_token_id,
        temperature=0.8,  # Adds randomness
        top_p=0.9,  # Nucleus sampling
        repetition_penalty=1.2  # Penalize repetitive phrases
    )
    return tokenizer.decode(outputs[0], skip_special_tokens=True)

def conversation_loop(start_message):
    conversation_history = [f"Agent 1: {start_message}"]  # Initialize conversation history
    print(f"Agent 1: {start_message}")

    for _ in range(5):  # 5 rounds of conversation
        # Prepare Agent 2's prompt
        agent2_prompt = "\n".join(conversation_history[-2:]) + "\nAgent 2:"
        agent2_message = get_response(agent2_prompt).strip()
        print(f"Agent 2: {agent2_message}")
        conversation_history.append(f"Agent 2: {agent2_message}")

        # Prepare Agent 1's prompt
        agent1_prompt = "\n".join(conversation_history[-2:]) + "\nAgent 1:"
        agent1_message = get_response(agent1_prompt).strip()
        print(f"Agent 1: {agent1_message}")
        conversation_history.append(f"Agent 1: {agent1_message}")

        # Limit conversation history to avoid excessive context
        if len(conversation_history) > 6:
            conversation_history = conversation_history[-6:]

# Start the conversation
conversation_loop("What do you know about AI?")
```

```python
from transformers import AutoModelForCausalLM, AutoTokenizer

# Load model and tokenizer
model_name = "gpt2"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)

def get_response(prompt):
    # Tokenize the input
    inputs = tokenizer(prompt, return_tensors="pt", truncation=True, max_length=200)
    outputs = model.generate(
        **inputs,
        max_new_tokens=50,
        pad_token_id=tokenizer.eos_token_id,
        temperature=0.8,  # Adds randomness
        top_p=0.9,  # Nucleus sampling
        repetition_penalty=1.2,  # Penalize repetitive phrases
        no_repeat_ngram_size=2  # Prevent n-gram repetition
    )
    return tokenizer.decode(outputs[0], skip_special_tokens=True)

def conversation_loop(start_message):
    conversation_history = [f"Agent 1: {start_message}"]  # Initialize conversation history
    print(f"Agent 1: {start_message}")

    for _ in range(10):  # 5 rounds of conversation
        # Prepare Agent 2's prompt (using the last two exchanges)
        agent2_prompt = "\n".join(conversation_history[-2:]) + "\nAgent 2:"
        agent2_message = get_response(agent2_prompt).strip()
        print(f"Agent 2: {agent2_message}")
        conversation_history.append(f"Agent 2: {agent2_message}")

        # Prepare Agent 1's prompt (using the last two exchanges)
        agent1_prompt = "\n".join(conversation_history[-2:]) + "\nAgent 1:"
        agent1_message = get_response(agent1_prompt).strip()
        print(f"Agent 1: {agent1_message}")
        conversation_history.append(f"Agent 1: {agent1_message}")

        # Limit conversation history to avoid excessive context
        if len(conversation_history) > 8:
            conversation_history = conversation_history[-8:]

# Start the conversation
conversation_loop("What do you know about AI?")
```
